```python
import os
import requests
import time

def download_arxiv_papers_kaggle(paper_ids, save_directory=r"C:\Users\HP EliteBook 840 G6\Desktop\TEST DATA\target_id", delay=2):
    # Create the directory if it doesn't exist
    os.makedirs(save_directory, exist_ok=True)
    
    # Loop through each paper ID and download the corresponding paper
    for paper_id in paper_ids:
        url = f"https://arxiv.org/pdf/{paper_id.strip()}"
        response = requests.get(url)
        
        if response.status_code == 200:
            # Save the file with the paper ID as the filename
            file_path = os.path.join(save_directory, f"{paper_id.strip()}.pdf")
            with open(file_path, 'wb') as f:
                f.write(response.content)
            print(f"Downloaded: {paper_id.strip()}")
        else:
            print(f"Failed to download: {paper_id.strip()} (Status code: {response.status_code})")
        
        # Wait for the specified delay before the next request
        time.sleep(delay)
```

```python
import os
import pandas as pd
from lxml import etree

# Folder containing your XML files
folder_path = r'C:\Users\HP EliteBook 840 G6\Desktop\TEST DATA\pdfsout'
# Function to extract ID from XML filename
def extract_id_from_filename(filename):
    return filename.replace('.grobid.tei.xml', '')
# Function to extract text content from XML file, excluding formulas, references, etc.
def extract_text_from_xml(file_path):
    with open(file_path, 'rb') as file:  # Open file in binary mode to avoid encoding issues
        xml_content = file.read()
    tree = etree.fromstring(xml_content)
    namespaces = {'tei': 'http://www.tei-c.org/ns/1.0'}
    # Extract the <body> content, excluding formulas, references, etc.
    body = tree.xpath('//tei:text/tei:body//tei:p', namespaces=namespaces)
    
    text_content = []
    for p in body:
        if not p.xpath('.//tei:formula | .//tei:ref', namespaces=namespaces):
            text_content.append(p.text)
    # Join the text content
    return '\n'.join([text for text in text_content if text])
# Initialize list to hold data
data = []
# Iterate over all XML files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.grobid.tei.xml'):
        paper_id = extract_id_from_filename(filename)
        file_path = os.path.join(folder_path, filename)
        text = extract_text_from_xml(file_path)     
        # Add extracted data to the list
        data.append({'ID_tar': paper_id, 'Text_tar': text})

# Create a DataFrame
target_data = pd.DataFrame(data)

# Save the DataFrame to a CSV file
output_path = r'C:\path\to\output\extracted_texts.csv'
#df.to_csv(output_path, index=False)

print(f"Dataset saved to {output_path}")
```

```python
from transformers import AutoModel, AutoTokenizer
# Load SciBERT
tokenizer = AutoTokenizer.from_pretrained("allenai/scibert_scivocab_uncased")
model = AutoModel.from_pretrained("allenai/scibert_scivocab_uncased")
```

```python
import torch
def get_embeddings(text):
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=512)
    with torch.no_grad():
        outputs = model(**inputs)
    # Use the [CLS] token's output as the embedding
    embeddings = outputs.last_hidden_state[:, 0, :].numpy()
    return embeddings
```
